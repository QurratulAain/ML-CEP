```python
dataset = dataset.dropna()
```

```python
# Count the number of occurrences of each class
dataset['Summary'].value_counts()
```

```python
# Selecting the top 4 most frequent classes
# Find top 4 classes
top_classes = dataset['Summary'].value_counts().index[:4].tolist()
# Create an empty list to store the indices of rows to keep
rows_to_keep = []

# Loop through each row in the dataframe
for index, row in dataset.iterrows():
    # Check if the 'Summary' value is in the top 4 classes
    if row['Summary'] in top_classes:
        rows_to_keep.append(index)

# Filter the DataFrame based on the indices to keep
dataset_filtered = dataset.loc[rows_to_keep]
print(dataset_filtered['Summary'].value_counts())
```

### splitting the dataSet

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2,random_state=42)
```

# KNN Model

```python
# Method 1: Count the occurrences of each class
class_counts = Y.value_counts()
print(class_counts)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
from sklearn import metrics 
from sklearn.metrics import confusion_matrix, classification_report
# Loading the knn classifier
classifier1 = KNeighborsClassifier(n_neighbors=7)
# Training the model
model1=classifier1.fit(X_train, Y_train)
# Evaluating the model
Y_pred1 = model1.predict(X_test) 
print("The accuracy is "+str(metrics.accuracy_score(Y_test,Y_pred1)*100)+"%") 
print(confusion_matrix(Y_test, Y_pred1))
target_names = ['class 0', 'class 1', 'class 2', 'class 3'] 
print(classification_report(Y_test, Y_pred1, target_names=target_names)) 
```

```python
# Using scikit-learn's weighted k-NN
from sklearn.neighbors import KNeighborsClassifier

# 'distance' weighting gives more influence to closer neighbors
weighted_knn = KNeighborsClassifier(n_neighbors=13, weights='distance', algorithm='auto')
weighted_knn.fit(X_train, Y_train)
Y_pred = weighted_knn.predict(X_test)
print("The accuracy is "+str(metrics.accuracy_score(Y_test,Y_pred)*100)+"%") 
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# Define the parameter grid
param_grid = {
    'n_neighbors': [5, 7, 9, 11, 13, 15],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
}
# Use stratified k-fold to maintain class distribution
stratified_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Create and fit the grid search
knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, param_grid, cv=stratified_cv, scoring='balanced_accuracy')
grid_search.fit(X_train, Y_train)

# Print all best parameters
print(f"Best parameters: {grid_search.best_params_}")
# # Or print each parameter individually
# print(f"Best k: {grid_search.best_params_['n_neighbors']}")
# print(f"Best weights: {grid_search.best_params_['weights']}")
# print(f"Best algorithm: {grid_search.best_params_['algorithm']}")
best_knn = grid_search.best_estimator_
```
